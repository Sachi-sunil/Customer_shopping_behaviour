```python
import pandas as pd
```

```python
df = pd.read_csv('Chennai_Clothing_Shop_Data.csv',sep="\t")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
#for reviews missing
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.rename(columns={
    "purchase_amount_(inr)": "purchase_amount",
}, inplace=True)
```

```python
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
#4 equal-sized quartiles and assigns each record one of the labels
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Monthly' : 30,
    'Weekly' : 7,
    'Fortnightly' : 14,
    'Occasionally': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']]
```

```python
df[['discount_applied','promo_code_used']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```

```python
df['location'] = df['location'].str.replace('Chennai, ', '', regex=False)
df['location'].unique()
```
